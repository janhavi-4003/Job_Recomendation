```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Job Datsset.csv")
df
```

```python
print(df.info)
```

```python
# Step 3: Check basic info
print("Initial Data Overview:")
print(df.shape)
print(df.columns)
print(df.isnull().sum())

```

```python
# Step 4: Drop rows with missing values (optional, if any)
df.dropna(inplace=True)

```

```python
# Step 5: Clean 'User_Skills' and 'Job_Requirements' text columns
def clean_text(text):
    return ",".join([i.strip().lower() for i in text.split(",")])

df["User_Skills"] = df["User_Skills"].apply(clean_text)
df["Job_Requirements"] = df["Job_Requirements"].apply(clean_text)

```

```python
# Step 6: Check unique values and data types (optional)
print("\nAfter cleaning:")
print(df.dtypes)
print(df.head(2))
```

```python
# Step 7: Save the cleaned data
df.to_csv("cleaned_data.csv", index=False)
print("\nâœ… Cleaned data saved to 'cleaned_data.csv'")
```
